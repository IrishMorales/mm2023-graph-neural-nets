The math behind a very basic network would look like below. Try to interpret what's going on in this diagram before reading the explanation below :>

(For this whole diagram, pretend $b=1$ and the activation function is ReLU.)

image_2023-10-02_033014755.png

Explanation:

1.   Input layer has one input, $5$.
2.   Hidden layer has one node. The computation for this node is: $activationFunction((input \cdot inputWeight) + (bias \cdot biasWeight))$.

If we break this down this means:

*   input = 5
*   input weight = 0.1
*   bias = b - let's pretend $b = 1$
*   bias weight = 0.2
*   activation function - let's pretend it's ReLU. If you see the image above, ReLU = $max(0, x)$

So, for this node, the computation would be

*   $output = max(0, ((5 \cdot 0.1) + (1 \cdot 0.2)))$
*   $output = max(0, (0.5 + 0.2))$
*   $output = max(0, (0.7))$
*   $output = 0.7$

3.   Output layer has one node. The computation for this node is the same as the hidden layer :D I'll let you do the math for this one, but to check your solution, the output for this node should be $0.61$.
