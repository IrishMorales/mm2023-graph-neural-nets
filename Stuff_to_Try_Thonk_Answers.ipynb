*   What information will node B have after 2 rounds of message passing?

- **Node B will have information from all other nodes**
*   What's the farthest node that node C will have information from?

- **The farthest node that node C will have information from is node A, since node A is 2 edges away and we did 2 rounds of message passing.**

image.png

*  Is the aggregated data per node alone enough to learn from?

*  **No, we also need to feed in the adjacency matrix so we learn the information from each node (the aggregated data) + the connections between nodes (the adjacency matrix)**

* If you have a model with 2 GCN layers, how many edges does data travel away from its source node? - **2**
* Notice that after the 2nd layer, each node now has one value. What kinds of models usually have one value as output? - **Binary classification**

*   How are GNNs better than just feeding data from each node?

*   **Feeding data from each node = same as vanilla neural networks that don't learn connections between nodes (no connection information)**

*   How are GNNs better than having graphs with just one data/value/number per node?

*   **Feeding one data/value/number per node = loses richness of information from multiple features in one node**
